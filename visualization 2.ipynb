```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#importing the data
tips =sns.load_dataset('tips')
```

```python
tips
```

```python
#stripplot
sns.stripplot(data=tips,x="day",y="tip",jitter=True)
plt.ylabel("tip")
plt.show()
```

```python
sns.stripplot(data=tips,y="tip",jitter=True)
plt.ylabel("tip")
plt.show()
```

```python
sns.stripplot(data=tips,y="tip",x="day",jitter=False)
plt.ylabel("tip")
plt.show()
```

```python
#swarmplot
sns.swarmplot(data=tips,y="tip",x="day")
plt.ylabel("tip")
plt.show()
```

```python
sns.swarmplot(x='tip', y='day', data=tips, hue='sex',orient='h')
plt.xlabel('tip ($)')
plt.show()
```

```python

sns.swarmplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tip ($)')
plt.show()
```

```python
#boxplot
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
```

```python
#violinplot
sns.violinplot(x="day",y="tip",data=tips)
plt.ylabel("tips")
plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x="day",y="tip",data=tips)
plt.ylabel("tips")
plt.subplot(1,2,2)
sns.violinplot(x="day",y="tip",data=tips)
plt.ylabel("tips")
plt.show()
```

```python
#combining stripplot and violinplot
sns.violinplot(x='day', y='tip', data=tips, inner=None,color='lightgray')
sns.stripplot(x='day', y='tip', data=tips, size=4,jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

```python
#jointplot
sns.jointplot(x="total_bill",y="tip",data=tips)
plt.show()
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="kde")
plt.show()
```

```python
#pairplot
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips,hue="sex")
plt.show()
```
